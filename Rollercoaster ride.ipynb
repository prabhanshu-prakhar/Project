```python
df = df[['coaster_name',  'Location', 'Status',  'Manufacturer',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 'speed_mph', 'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
pd.to_numeric(df['year_introduced'])
```

```python
df = df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'})
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])] \
    .reset_index(drop=True).copy()
```

# feature engineering

```python
df['Year_Introduced'].value_counts().head(10).plot(kind = 'bar')

plt.xlabel('Year Introduced')
plt.ylabel('counts')
```

```python
df['Speed_mph'].plot(kind= 'hist', bins = 20)
```

```python
df['Speed_mph'].plot(kind= 'kde', title = 'coaster speed')
```

```python
sns.scatterplot(data = df, x = 'Speed_mph', y = 'Height_ft', hue = 'Year_Introduced' )
plt.xlabel('Height')
plt.ylabel('Speed')
plt.title('Speed verses height')

plt.figure(figsize=(12, 6))
plt.show()
```

```python
sns.pairplot(df, vars = ['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions','Gforce'], hue = 'Type_Main')
plt.show()
```

```python
df_corr = df[ ['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions','Gforce']].dropna().corr()
```

```python
df_corr
```

```python
sns.heatmap(df_corr, cmap = 'coolwarm', annot = True)
```

```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
